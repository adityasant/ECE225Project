Extracting the temperature data for USA from 1973 - 2015

```python
start_year = 1973
start_date = str(start_year)+'-01-01'
end_year = 2012
end_date = str(end_year)+'-12-01'
country = 'United States'
num_years = end_year - start_year + 1
```

```python
temp_data_USA = temp_data0[temp_data0['Country'].str.contains(country)]
temp_data = []
date_data = []
st = 0
end = 0
ind = 0
for d in temp_data_USA['dt']:
    if(d==start_date):
        st = ind
    if(st!=0 and end==0):
        date_data.append(d)
    if(d==end_date):
        end = ind
    ind+=1
```

Creating a linear fit model to the 12 months

```python
x_arr = np.linspace(0,num_years-1,num_years)
p = np.zeros([12,2])
```

```python
for k in range(12):
    p[k,:] = np.polyfit(x_arr,t1[:,k],1)
```

```python
plt.figure()
plt.plot(x_arr,t1[:,0])
plt.plot(x_arr,np.polyval(p[0,:],x_arr))
plt.plot(x_arr,t1[:,5])
plt.plot(x_arr,np.polyval(p[5,:],x_arr))
```
